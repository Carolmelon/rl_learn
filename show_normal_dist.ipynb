```python
import matplotlib.pyplot as plt

# 定义一系列点的坐标
x = [2,3,4]  # x坐标列表
y = [1,2,5]  # y坐标列表

# 创建散点图
plt.scatter(x, y)

# 设置图形的标题和坐标轴标签
plt.title("Scatter Plot")
plt.xlabel("X")
plt.ylabel("Y")

# 显示图形
plt.show()
```

```python
import torch
from torch.distributions import Normal

# 创建均值为0、标准差为1的正态分布对象
mean = torch.tensor(0.0)
std = torch.tensor(5)
normal_dist = Normal(mean, std)
```

```python
# 生成随机样本
sample = normal_dist.sample()
print(sample)
```

```python
data = []
n_points = 100000
for i in range(n_points):
    data.append(normal_dist.sample().item())
```

```python
len(data)
```

```python
# 绘制频率密度图
plt.hist(data, density=True, bins=100)  # density=True表示绘制的是频率密度而不是频数
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Density Plot')

# 显示图形
plt.show()
```
